```python
X=df.drop(columns='TARGET',axis=1)
Y=df['TARGET']
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
model1=LogisticRegression()
model1.fit(X_train,Y_train)
X_train_prediction=model1.predict(X_train)
training_data_accuarcy=accuracy_score(X_train_prediction,Y_train)
print('Accuarcy on training data:',training_data_accuarcy)
X_test_prediction=model1.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print('Accuarcy on test data:',test_data_accuarcy)
```

```python
model2=DecisionTreeClassifier()
model2.fit(X_train,Y_train)
X_train_prediction=model2.predict(X_train)
training_data_accuarcy=accuracy_score(X_train_prediction,Y_train)
print('Accuarcy on training data:',training_data_accuarcy)
X_test_prediction=model1.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print('Accuarcy on test data:',test_data_accuarcy)
```

```python
Model3=RandomForestClassifier()
Model3.fit(X_train,Y_train)
X_train_prediction=Model3.predict(X_train)
training_data_accuarcy=accuracy_score(X_train_prediction,Y_train)
print('Accuarcy on training data:',training_data_accuarcy)
X_test_prediction=Model3.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print('Accuarcy on test data:',test_data_accuarcy)
```

```python
Model4=KNeighborsClassifier(n_neighbors=3)
Model4.fit(X_train,Y_train)
X_train_prediction=Model4.predict(X_train)
training_data_accuarcy=accuracy_score(X_train_prediction,Y_train)
print('Accuarcy on training data:',training_data_accuarcy)
X_test_prediction=Model4.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print('Accuarcy on test data:',test_data_accuarcy)
```

```python
Model5=SVC()
Model5.fit(X_train,Y_train)
X_train_prediction=Model5.predict(X_train)
training_data_accuarcy=accuracy_score(X_train_prediction,Y_train)
print('Accuarcy on training data:',training_data_accuarcy)
X_test_prediction=Model5.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print('Accuarcy on test data:',test_data_accuarcy)
```

```python
Model6=GaussianNB()
Model6.fit(X_train,Y_train)
X_train_prediction=Model6.predict(X_train)
training_data_accuarcy=accuracy_score(X_train_prediction,Y_train)
print('Accuarcy on training data:',training_data_accuarcy)
X_test_prediction=Model6.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print('Accuarcy on test data:',test_data_accuarcy)
```

```python
Modle7=xgb.XGBClassifier(random_state=1)
Modle7.fit(X_train,Y_train)
X_train_prediction=Modle7.predict(X_train)
training_data_accuarcy=accuracy_score(X_train_prediction,Y_train)
print('Accuarcy on training data:',training_data_accuarcy)
X_test_prediction=Modle7.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print('Accuarcy on test data:',test_data_accuarcy)
```
